```python
# Đặt phạm vi (scope)
scope = ["https://spreadsheets.google.com/feeds", "https://www.googleapis.com/auth/drive"]

# Tải credentials từ file JSON
creds = ServiceAccountCredentials.from_json_keyfile_name('/content/drive/MyDrive/Dự án/Báo cáo outsource anh Tuấn/service_account_credentials.json', scope)
client = gspread.authorize(creds)

# Mở Google Sheet bằng URL hoặc ID
spreadsheets = client.open_by_url('https://docs.google.com/spreadsheets/d/1fA6v3fc_hOCAFod0rbob_Edv3_2J6gW3GlRf0Vib4Rc/edit?gid=1322719953#gid=1322719953')
ki_nop_phat_sheet = spreadsheets.get_worksheet_by_id('1322719953').get_all_records()
Nop_phat_thuc_te_sheet = spreadsheets.get_worksheet_by_id('1303177755').get_all_records()
```

```python
# lấy bảng kì hạn dự kiến
ki_nop_phat_df = pd.DataFrame(ki_nop_phat_sheet)
ki_nop_phat_df
```

```python
ki_nop_phat_df['Ngày đến hạn'] = pd.to_datetime(ki_nop_phat_df['Ngày đến hạn'], format='%Y-%m-%d')
```

```python
#lấy bảng nộp phạt thực tế
nop_phat_thuc_te_df = pd.DataFrame(Nop_phat_thuc_te_sheet).iloc[:,:5]
nop_phat_thuc_te_df
```

```python
nop_phat_thuc_te_df["Ngày vào sổ"] = pd.to_datetime(nop_phat_thuc_te_df['Ngày vào sổ'], format='%m/%d/%Y')
```

```python
nop_phat_thuc_te_grb_df  = nop_phat_thuc_te_df.groupby(["Mã hợp đồng","Kỳ đóng tiền"]).agg({
    'Thời gian trả': 'min',
    'Số tiền nộp phạt': 'sum'}).reset_index()
nop_phat_thuc_te_grb_df
```

```python
# Merge
merged_df = pd.merge(nop_phat_thuc_te_df, nop_phat_thuc_te_grb_df, on=['Mã hợp đồng','Kỳ đóng tiền'], suffixes=('', '_grouped'))

# Loại bỏ các hàng trùng lặp dựa trên 'Kỳ đóng tiền'
final_nop_phat_thuc_te_grb_df = merged_df.drop_duplicates(['Mã hợp đồng','Kỳ đóng tiền'])
```

```python
final_nop_phat_thuc_te_grb_df=final_nop_phat_thuc_te_grb_df.reset_index().drop(columns=["Thời gian trả","Số tiền nộp phạt",'index'])
```

```python
final_df = pd.merge(ki_nop_phat_df,final_nop_phat_thuc_te_grb_df, how='left',left_on=['Mã hợp đồng', 'Kỳ hạn'],right_on=["Mã hợp đồng","Kỳ đóng tiền"])
```

```python
final_df["Ngày quá hạn"] = (final_df['Ngày vào sổ'] - final_df['Ngày đến hạn']).dt.days
```

```python
final_df["Ngày quá hạn"] = final_df["Ngày quá hạn"].fillna(20).astype(int)
```

```python
final_df["Số tiền nộp phạt_grouped"] = final_df["Số tiền nộp phạt_grouped"].fillna(0).astype(int)
```

```python
final_df["Tiền phạt kỳ này"] = final_df["Ngày quá hạn"].apply(lambda  x: (x-5)*100000 if x>6 else 0)
```

```python
final_df=final_df.sort_values(by=["Mã hợp đồng","Kỳ hạn"]).reset_index()
```

```python
final_df.loc[:,["Mã hợp đồng","Số tiền nộp phạt_grouped","Kỳ hạn","Tiền phạt kỳ này"]]
```

```python
# tiền phạt lũy kế cuối kỳ >= 0 và <= 2000000
def checkdieukiensotienphat(a):
    x = 0
    # if a < 0:
    #     x = 0
    if a > 2000000:
        x = 2000000
    else:
        x = a
    return x
```

```python
# Tính Tiền phạt lũy kế cuối kỳ
final_df['Tiền phạt lũy kế cuối kỳ'] = 0
for contract_id in final_df['Mã hợp đồng'].unique():
    mask = final_df['Mã hợp đồng'] == contract_id
    first_index = final_df[mask]['Kỳ hạn'].idxmin()
    first_value = final_df.loc[first_index, 'Tiền phạt kỳ này']
    final_df.loc[first_index, 'Tiền phạt lũy kế cuối kỳ'] = first_value

    for i in range(first_index + 1, len(final_df)):
        if final_df.loc[i, 'Mã hợp đồng'] == contract_id:
            previous_cumulative = final_df.loc[i - 1, 'Tiền phạt lũy kế cuối kỳ']
            current_penalty = final_df.loc[i, 'Tiền phạt kỳ này']
            current_grouped_penalty = final_df.loc[i, 'Số tiền nộp phạt_grouped']
            cumulative_penalty = current_penalty + previous_cumulative - current_grouped_penalty
            cumulative_penalty = checkdieukiensotienphat(cumulative_penalty)

            final_df.loc[i, 'Tiền phạt lũy kế cuối kỳ'] = cumulative_penalty
```
