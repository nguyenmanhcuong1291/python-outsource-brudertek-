```python
# lấy bảng kì hạn dự kiến
ki_nop_phat_sheet = spreadsheets.get_worksheet_by_id('1322719953').get_all_records()
ki_nop_phat_df = pd.DataFrame(ki_nop_phat_sheet)
ki_nop_phat_df
```

```python
ki_nop_phat_df['Ngày đến hạn'] = pd.to_datetime(ki_nop_phat_df['Ngày đến hạn'], format='%Y-%m-%d')
```

```python
#lấy bảng nộp phạt thực tế
Nop_phat_thuc_te_sheet = spreadsheets.get_worksheet_by_id('1303177755').get_all_records()
nop_phat_thuc_te_df = pd.DataFrame(Nop_phat_thuc_te_sheet).iloc[:,:5]
nop_phat_thuc_te_df
```

```python
nop_phat_thuc_te_df['Ngày vào sổ'] = pd.to_datetime(nop_phat_thuc_te_df['Ngày vào sổ'], format='%Y-%m-%d')
```

```python
giam_tru_sheet = spreadsheets.get_worksheet_by_id('1700047741').get_all_records()
```

```python
# lấy bảng kì hạn dự kiến
giam_tru_df = pd.DataFrame(giam_tru_sheet)
giam_tru_df
```

```python
nop_phat_thuc_te_grb_df  = nop_phat_thuc_te_df.groupby(["Mã hợp đồng","Kỳ đóng tiền"]).agg({
    'Thời gian trả': 'min',
    'Tổng tiền đã đóng': 'sum'}).reset_index()
nop_phat_thuc_te_grb_df
```

```python
# Merge
merged_df = pd.merge(nop_phat_thuc_te_df, nop_phat_thuc_te_grb_df, on=['Mã hợp đồng','Kỳ đóng tiền'], suffixes=('', '_grouped'))

# Loại bỏ các hàng trùng lặp dựa trên 'Kỳ đóng tiền'
final_nop_phat_thuc_te_grb_df = merged_df.drop_duplicates(['Mã hợp đồng','Kỳ đóng tiền'])
```

```python
final_nop_phat_thuc_te_grb_df=final_nop_phat_thuc_te_grb_df.reset_index().drop(columns=["Thời gian trả","Tổng tiền đã đóng",'index'])
```

```python
# tiền phạt lũy kế cuối kỳ >= 0 và <= 2000000
def checkdieukiensotienphat(a):
    x = 0
    # if a < 0:
    #     x = 0
    if a > 2000000:
        x = 2000000
    else:
        x = a
    return x
```

```python
def check_tienthu_goc(a,b):
  if a<0:
    x = 0
  else:
      if a>=b:
        x = b
      else:
        x = a
  return x


```

```python
for mahopdong in final_df['Mã hợp đồng'].unique():
    partition_by_mahopdong = final_df['Mã hợp đồng'] == mahopdong
    first_index = final_df[partition_by_mahopdong]['Kỳ hạn'].idxmin()
    first_phat_luy_ke = final_df.loc[first_index, 'Tiền phạt kỳ này']
    first_goc_dadong =  check_tienthu_goc( final_df.loc[first_index, 'Tổng tiền đã đóng'], final_df.loc[first_index, 'Gốc kế hoạch'])
    final_df.loc[first_index, 'Gốc đã đóng'] = first_goc_dadong
    first_lai_dadong = check_tienthu_goc( final_df.loc[first_index, 'Tổng tiền đã đóng']-first_goc_dadong, final_df.loc[first_index, 'Lãi kế hoạch'])
    final_df.loc[first_index, 'Lãi đã đóng'] = first_lai_dadong
    first_phat_dadong = check_tienthu_goc(final_df.loc[first_index, 'Tổng tiền đã đóng']-first_goc_dadong-first_lai_dadong,final_df.loc[first_index, 'Tiền phạt kỳ này'])
    final_df.loc[first_index, 'Phạt đã đóng'] = first_phat_dadong
    final_df.loc[first_index, 'Gốc nợ kỳ trước'] = 0
    final_df.loc[first_index, 'Lãi nợ kỳ trước'] = 0
    final_df.loc[first_index, 'Phạt nợ kỳ trước'] = 0
    # final_df.loc[first_index, ] = final_df.loc[first_index, 'Gốc kế hoạch'] - first_goc_dadong
    # final_df.loc[first_index, 'Lãi cuối kỳ'] = final_df.loc[first_index, 'Lãi kế hoạch'] - first_lai_dadong
    # final_df.loc[first_index, 'Tiền phạt lũy kế cuối kỳ'] = final_df.loc[first_index, 'Tiền phạt kỳ này'] - first_lai_dadong
    # final_df.loc[first_index, 'Tiền phạt nợ kỳ trước'] = 0
    for i in range(first_index + 1, len(final_df)):
      if final_df.loc[i, 'Mã hợp đồng'] == mahopdong:
        final_df.loc[i, 'Gốc nợ kỳ trước'] = final_df.loc[i-1, 'Gốc kế hoạch'] +  final_df.loc[i-1, 'Gốc nợ kỳ trước'] - final_df.loc[i-1, 'Gốc đã đóng']
        final_df.loc[i, 'Lãi nợ kỳ trước'] = final_df.loc[i-1, 'Lãi kế hoạch'] +  final_df.loc[i-1, 'Lãi nợ kỳ trước'] - final_df.loc[i-1, 'Lãi đã đóng']
        final_df.loc[i, 'Phạt nợ kỳ trước'] = final_df.loc[i-1, 'Tiền phạt kỳ này'] +  final_df.loc[i-1, 'Phạt nợ kỳ trước'] - final_df.loc[i-1, 'Phạt đã đóng']
        final_df.loc[i, 'Gốc đã đóng'] = check_tienthu_goc( final_df.loc[i, 'Tổng tiền đã đóng'], final_df.loc[i, 'Gốc kế hoạch'] + final_df.loc[i, 'Gốc nợ kỳ trước'])
        final_df.loc[i, 'Lãi đã đóng'] = check_tienthu_goc( final_df.loc[i, 'Tổng tiền đã đóng'] - final_df.loc[i, 'Gốc đã đóng'], final_df.loc[i, 'Lãi kế hoạch'] + final_df.loc[i, 'Lãi nợ kỳ trước'])
        final_df.loc[i, 'Phạt đã đóng'] = check_tienthu_goc( final_df.loc[i, 'Tổng tiền đã đóng'] - final_df.loc[i, 'Gốc đã đóng']- final_df.loc[i, 'Lãi đã đóng'], checkdieukiensotienphat(final_df.loc[i, 'Tiền phạt kỳ này'] + final_df.loc[i, 'Phạt nợ kỳ trước']))
```
